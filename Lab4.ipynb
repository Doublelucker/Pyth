It's not really a good solution. But if we turn the shuffling of the data off, then we can achieve almost 100% accuracy. Assigning of the labels is still random, but if we sort the labels, then they align with the y.

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=42, shuffle = False)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
```

```python
fig, axes = plt.subplots(2, figsize = (6,9))
axes[0].scatter(X[:, 0], X[:, 1], c = y);
axes[1].scatter(X[:, 0], X[:, 1], c=np.sort(labels));
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print('Accuracy score:', accuracy_score(y, np.sort(labels)))
print(confusion_matrix(y, np.sort(labels)))
```

```python
np.mean(y == np.sort(labels))
```

Even though we recovered the partitioning of the data into clusters perfectly, the cluster IDs we assigned were arbitrary,
and we can not hope to recover them. Therefore, we must use a different scoring metric, such as ``adjusted_rand_score``, which is invariant to permutations of the labels:

```python
from sklearn.metrics import adjusted_rand_score

adjusted_rand_score(y, labels)
```

### 1. Training

```python
from sklearn.datasets import load_digits
from sklearn.preprocessing import scale
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.cluster import MeanShift
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import SpectralClustering

digits = load_digits()
dig = scale(digits.data)
n_digits = len(np.unique(digits.target))
labels = digits.target
n_samples, n_features = dig.shape

re_data = TSNE(n_components = 2, perplexity = 40, learning_rate = 500,
            n_iter = 500, random_state = 32).fit_transform(dig)
km = KMeans(init = 'k-means++', n_clusters = n_digits, n_init = 10)
meanshift = MeanShift(bin_seeding = True)
ap = AffinityPropagation(damping = .7, convergence_iter = 10, max_iter = 800,
                        preference=-1100)
ac = []
for linkage in ('ward', 'average', 'complete'):
    ac.append(AgglomerativeClustering(n_clusters = n_digits,linkage = linkage))
sc = SpectralClustering(n_clusters = n_digits, n_init = 10, degree = 5, gamma = 0.01, random_state = 32)
    
sc.fit(re_data)    
ap.fit(re_data)
km.fit(re_data)
meanshift.fit(re_data)
```

```python
def plot_clustering(X_red, classifier, title=None):
    plt.figure(figsize=(6, 4))
    #if (classifier not in [km, ap]):
    x_min, x_max = np.min(X_red, axis=0), np.max(X_red, axis=0)
    X_red = (X_red - x_min) / (x_max - x_min)
    for i in range(X_red.shape[0]):
        plt.text(X_red[i, 0], X_red[i, 1], str(labels[i]),
                 color=plt.cm.spectral(classifier.labels_[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})   
    #if (classifier in [km, ap]):
        #centr = classifier.cluster_centers_
        #for i in range(centr.shape[0]):
             #plt.scatter(centr[i,0], centr[i,1],color = 'pink', marker = '+', s=150, lw = 4, zorder = 10)
    plt.xticks([])
    plt.yticks([])
    if title is not None:
        plt.title(title, size=17)
    plt.axis('off')
    plt.tight_layout()
```
